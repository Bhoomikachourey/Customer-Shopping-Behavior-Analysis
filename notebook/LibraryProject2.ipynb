```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('student_data.csv')
df
```

# **add a new column**

```python
df['TotalScore'] = df['MathScore'] + df['ReadingScore'] + df['WritingScore']
df
```

# ***`DashBoard`***

```python
fig, axes=plt.subplots(2,3, figsize=(18,10),facecolor='#050D17')
plt.suptitle('Impact of Demographic and Lifestyle Factors on Student Academic Performance',fontsize=20,fontweight='bold',color='#FFD700')
#1. bar plot-> Gender vs Average Score
plot1=sns.barplot(x='Gender',y='TotalScore',data=df, ax=axes[0][0],hue='Gender',palette='Set1',errorbar=None)
axes[0][0].set_title('Overall Performance by Gender',color='#E0FFFF',fontweight='bold',fontsize=15)
axes[0][0].tick_params('x',colors='w',size=6,labelfontfamily='serif')
axes[0][0].tick_params('y',colors='w',size=6,labelfontfamily='serif')
axes[0][0].set_facecolor('#14253D')
axes[0][0].set_xlabel('Gender',color='cyan',fontsize=14)
axes[0][0].set_ylabel('Average Total Score',color='cyan',fontsize=14)
for i in plot1.containers:
  plot1.bar_label(i, color='white')

#2. bar plot-> parent education vs avg score
plot2=sns.barplot(y='ParentEduc', x='TotalScore', data=df,ax=axes[0][1],hue='ParentEduc',palette='tab10',errorbar=None)
axes[0][1].set_title('Impact of Parent Education',color='#E0FFFF',fontweight='bold',fontsize=15)
axes[0][1].tick_params('x',colors='w',size=6,labelfontfamily='serif')
axes[0][1].tick_params('y',colors='w',size=6,labelfontfamily='serif')
axes[0][1].set_facecolor('#14253D')
axes[0][1].set_xlabel('Average Total Score',color='cyan',fontsize=14)
axes[0][1].set_ylabel('Parent Education',color='cyan',fontsize=14)
for i in plot2.containers:
  plot2.bar_label(i, color='white',rotation=270)

#3. bar plot-> test preparation vs average scores
plot3=sns.barplot(y='TotalScore', x='TestPrep', data=df,ax=axes[0][2],hue='TestPrep',palette='Spectral',errorbar=None)
axes[0][2].set_title('Effect of Test Preparation',color='#E0FFFF',fontweight='bold',fontsize=15)
axes[0][2].tick_params('x',colors='w',size=6,labelfontfamily='serif')
axes[0][2].tick_params('y',colors='w',size=6,labelfontfamily='serif')
axes[0][2].set_facecolor('#14253D')
axes[0][2].set_xlabel('Average Total Score',color='cyan',fontsize=14)
axes[0][2].set_ylabel('Test Preparation',color='cyan',fontsize=14)
for i in plot3.containers:
  plot3.bar_label(i, color='white')

#4. bar plot-> weekly study hours vc avg score
plot4=sns.barplot(y='TotalScore', x='WklyStudyHours', data=df,ax=axes[1][0],hue='WklyStudyHours',palette='magma',errorbar=None)
axes[1][0].set_title('Impact of Weekly Study Hours',color='#E0FFFF',fontweight='bold',fontsize=15)
axes[1][0].tick_params('x',colors='w',size=6,labelfontfamily='serif')
axes[1][0].tick_params('y',colors='w',size=6,labelfontfamily='serif')
axes[1][0].set_facecolor('#14253D')
axes[1][0].set_xlabel('Weekly Study Hours',color='cyan',fontsize=14)
axes[1][0].set_ylabel('Average Total Score',color='cyan',fontsize=14)
for i in plot4.containers:
  plot4.bar_label(i, color='white')

#5. bar plot-> practice sport vc avg score
plot5=sns.barplot(y='TotalScore', x='PracticeSport', data=df,ax=axes[1][1],hue='PracticeSport',palette='Paired',errorbar=None)
axes[1][1].set_title('Effect of Sports Participation',color='#E0FFFF',fontweight='bold',fontsize=15)
axes[1][1].tick_params('x',colors='w',size=6,labelfontfamily='serif')
axes[1][1].tick_params('y',colors='w',size=6,labelfontfamily='serif')
axes[1][1].set_facecolor('#14253D')
axes[1][1].set_xlabel('Practice Sport',color='cyan',fontsize=14)
axes[1][1].set_ylabel('Average Total Score',color='cyan',fontsize=14, labelpad=15)
for i in plot5.containers:
  plot5.bar_label(i, color='white')

#6. heatmap-> mathscore, reading score, writing score
c_var=df[['MathScore',	'ReadingScore',	'WritingScore']].corr()
heatmap=sns.heatmap(c_var,cmap='Spectral',annot=True)
axes[1][2].set_title('Correlation b/w Math,Reading,and Writing Scores',color='#E0FFFF',fontweight='bold',fontsize=15)
axes[1][2].tick_params('x',colors='w',size=6,labelfontfamily='serif')
axes[1][2].tick_params('y',colors='w',size=6,labelfontfamily='serif')
axes[1][2].set_facecolor('#14253D')


# Change color bar tick labels to white
cbar = heatmap.collections[0].colorbar
cbar.ax.yaxis.set_tick_params(color='white')
plt.setp(cbar.ax.get_yticklabels(), color='white')

axes[0][1].legend().remove()
axes[0][2].legend().remove()
axes[1][0].legend().remove()
axes[1][1].legend().remove()

plt.tight_layout()
plt.savefig('DashBoard.png',dpi=1000)
plt.show()
```

# ***`Conclusion`***

**Gender Analysis:** The analysis shows that (Male/Female) students achieved a higher average overall score than the other gender.

**Parent Education:** Students whose parents have a higher level of education generally perform better academically, indicating a positive relationship between parental education and student achievement.

**Test Preparation:** Students who completed the test preparation course scored higher on average, suggesting that preparation contributes to better academic performance.

**Weekly Study Hours:** Students who spend more hours studying each week tend to achieve higher overall scores, highlighting the importance of consistent study habits.

**Sports Participation:** Students who regularly participate in sports show (better/similar/lower) academic performance compared to those who do not, based on the dataset.

**Correlation Analysis:** Math, Reading, and Writing scores are positively correlated, indicating that students who perform well in one subject are likely to perform well in the others.

# Overall Insight

This analysis indicates that parental education, test preparation, weekly study hours, and sports participation can influence student performance. Among these factors, test preparation and study hours show a strong positive association with higher academic scores. Additionally, comparing average total scores reveals which gender performs better overall in this dataset.
